```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #data visualizing
%matplotlib inline
import seaborn as sns 
```

```python
df= pd.read_csv('Diwali Sales Data.csv',encoding= 'unicode_escape' )
```

```python
df.shape  #show rows and column in the using data set 
```

```python
df.head() #show the column name of the data sets 
```

```python
df.info() #show the information of datasets
```

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace = True) # use to delete the columns  axis=1 means the complete vertical row implace=True means save in same line 
```

```python
pd.isnull(df).sum()  #check for null value 
```

```python
df.shape
```

```python
df.dropna(inplace=True)   # is use to delete the null value
```

```python
pd.isnull(df).sum()
```

```python
df['Amount']= df ['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.describe() #this method returns descriptions of the data in the dataframe 
```

```python
df[['Age', 'Orders', 'Amount']].describe()   # description of specific columns 
```

```python
df.columns
```

```python
ax = sns.countplot(x='Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender', y= 'Amount', data = sales_gen)
```

# from above graph we can see that most of the buyers are females and even the purchasing power of females are greater then man

# State

```python
df.columns
```

```python
#total  number of orders in state
df.groupby (['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
```

```python
sales_order = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(8)

sns.barplot(data = sales_order, x = 'State', y = 'Orders')

sns.set(rc= {'figure.figsize':(15,5)})
```

```python
sales_order = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(8)

sns.barplot(data = sales_order, x = 'State', y = 'Amount')

sns.set(rc= {'figure.figsize':(15,5)})
```

# from the above graph we can see that most of the order are from these states and highest order amount are also spents by these state female

# Marital status 

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = 'Marital_Status') 

for bars in ax.containers:
    ax.bar_label(bars)
    
sns.set(rc= {'figure.figsize':(7,3)})
```

```python

df.groupby (['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_state = df.groupby (['Marital_Status', 'Gender'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)

sns.barplot(data = sales_state, x = 'Marital_Status', y = 'Orders', hue= 'Gender')

sns.set(rc= {'figure.figsize':(7,3)})
```

# from the  above graph we got that most of the buyers are married (female) and they are high purchasing 

# Occupations

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = 'Occupation') 

for bars in ax.containers:
    ax.bar_label(bars)
sns.set(rc= {'figure.figsize':(27,3)})    
```

```python

sales_occup = df.groupby (['Occupation', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(data = sales_occup, x = 'Occupation', y = 'Amount', hue= 'Gender')

sns.set(rc= {'figure.figsize':(15,3)})
```

```python

sales_occup = df.groupby (['Occupation', 'Gender'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.barplot(data = sales_occup, x = 'Occupation', y = 'Orders', hue= 'Gender')
```

# most of the buyers are female are working in IT-Sector, Healthcare, Aviations etc.

# Product category

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = 'Product_Category') 

for bars in ax.containers:
    ax.bar_label(bars)
sns.set(rc= {'figure.figsize':(27,5)}) 
```

```python
#product on the basis of amount
sales_occup = df.groupby (['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(data = sales_occup, x = 'Product_Category', y = 'Amount')
```

# from the graph we got it shows that most ordered product is food, Clothing & Apparel, Electronics & Gadgets 

```python
# product on the basis of numbers of order 
sales_occup = df.groupby (['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.barplot(data = sales_occup, x = 'Product_ID', y = 'Orders')
```

# Conclusion:

# married women age group 26 - 35 yrs from up, maharastra and karnataka  working in IT Sector, health Care and Aviation are more like to buy products like food, Clothing & Apparel, Electronics & Gadgets etc
